```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('cr_rf_sample_100k.csv')
display(df.head())
```

```python
df = df[df['trophy_bin'] == 'bin_4']
df["cards_list"] = df["cards_str"].apply(lambda x: x.split("|"))

df = df.drop("cards_str", axis=1)

df
```

```python
wincons = pd.read_csv('Wincons.csv')
wincons["card_id"] = wincons["card_id"].astype(str)
wincons
```

```python
wincon_list = wincons["card_id"].tolist()

# Convert all entries to strings
wincon_list = [str(x) for x in wincon_list]

# Remove any placeholder or invalid card ID
if '99999999' in wincon_list:
    wincon_list.remove('99999999')

# Check the result
wincon_list
```

```python
#getting rid of the decks without wincons
df["wincons_found"] = df["cards_list"].apply(
    lambda cards: [card for card in cards if card in wincon_list]
)
df["has_wincon"] = df["wincons_found"].apply(lambda x: len(x) > 0)
df_wincon = df[df["has_wincon"]]

df_wincon
```

```python
usage_counts = {}

for wc in wincon_list:
    # Count how many decks have this wincon
    usage_counts[wc] = df["cards_list"].apply(lambda cards: wc in cards).sum()

winrates = {}

for wc in wincon_list:
    # Filter decks containing this wincon
    decks_with_wc = df[df["cards_list"].apply(lambda cards: wc in cards)]
    # Average of label = winrate
    winrates[wc] = decks_with_wc["label"].mean()


wincon_stats = pd.DataFrame({
    "wincon_id": list(usage_counts.keys()),
    "usage_count": list(usage_counts.values()),
    "winrate": [winrates[wc] for wc in usage_counts.keys()]
})

# Sort by usage or winrate
wincon_stats = wincon_stats.sort_values(by="winrate", ascending=False)

wincon_stats_named = wincon_stats.merge(
    wincons, 
    left_on="wincon_id", 
    right_on="card_id",
    how="left"
)
wincon_stats_named = wincon_stats_named.drop(columns=["card_id", "id"])

wincon_stats_named
```

```python
top_wincons = wincon_stats_named.sort_values(by="winrate", ascending=False).head(5)
top_usage = wincon_stats_named.sort_values(by="usage_count", ascending=False).head(5)

top_wincons
```
